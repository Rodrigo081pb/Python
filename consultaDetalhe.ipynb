```python
import requests
from bs4 import BeautifulSoup

# URL base
url = 'https://pje.cloud.tjpe.jus.br/1g/ConsultaPublica/DetalheProcessoConsultaPublica/listView.seam'

# Parâmetros de consulta
params = {
    'ca': '7e5f64e9e23f5008f4e79610e65bc3e086b0dc9ad12445ce'
}

try:
    # Faça a requisição GET
    response = requests.get(url, params=params)

    # Verifique se a requisição foi bem-sucedida
    if response.status_code == 200:
        print("Requisição bem-sucedida!")

        # Analise o conteúdo HTML com BeautifulSoup
        soup = BeautifulSoup(response.text, 'html.parser')

        # Encontrar todos os elementos `propertyView` no HTML
        property_views = soup.find_all('div', class_='propertyView')

        # Dicionário para armazenar os resultados
        results = {}

        for view in property_views:
            # Extrair o nome e o valor de cada `propertyView`
            name_div = view.find('div', class_='name')
            value_div = view.find('div', class_='value')
            
            if name_div and value_div:
                # Extrair o texto do nome e do valor
                name = name_div.get_text(strip=True)
                value = value_div.get_text(strip=True)
                
                # Armazenar no dicionário, removendo o texto de `label` dos nomes
                if name:
                    key = name.replace('Número Processo', 'Número Processo')\
                              .replace('Data da Distribuição', 'Data da Distribuição')\
                              .replace('Classe Judicial', 'Classe Judicial')\
                              .replace('Assunto', 'Assunto')\
                              .replace('Jurisdição', 'Jurisdição')\
                              .replace('Órgão Julgador', 'Órgão Julgador')
                    
                    # Adicionar um formato específico para o "Assunto"
                    if key == 'Assunto':
                        # Substituir "\n" por uma nova linha
                        value = value.replace('\n', '<br>')  # Usa '<br>' para visualização HTML ou '\n' para texto
                        
                    results[key] = value
        
        # Exibir resultados com formatação específica para o Assunto
        for key, value in results.items():
            if key == 'Assunto':
                # Para o Assunto, adicionar uma linha em branco entre os itens
                print(f"{key}:")
                for item in value.split('<br>'):
                    print(f"    {item.strip()}")
            else:
                print(f"{key}: {value}")

    else:
        print("Falha na requisição. Código de status:", response.status_code)
        print("Resposta:", response.text)

except requests.RequestException as e:
    print("Erro na requisição:", e)
```
